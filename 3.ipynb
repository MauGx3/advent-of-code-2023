```python
with open('engine.txt', 'r') as f:
    engine_schematics = f.readlines()

engine_schematics = [line.strip() for line in engine_schematics]
gear_sum = 0
current_line = {'number_pos': [], 'number_match': [], 'symbol_pos': []}
print(len(engine_schematics))
engine_schematics
```

```python
import re


def find_numbers(line_number, current_line): #! might not work with numbers with len < 3
    global number_span, number_match
    for match in re.finditer(r'\d+', engine_schematics[line_number]):  # fixme
        number_span = list(match.span())
        number_match = match.group()
        number_span[1] = number_span[1] - 1
        number_span.insert(1, number_span[0] + 1)
        current_line['number_pos'] += [number_span]
        current_line['number_match'] += [int(number_match)]
        # print(number_span, number_match)
```

```python
def find_symbols(line_number, current_line):
    global symbol_pos, symbol_match
    for match in re.finditer(r'([^a-zA-Z\d\s:\.])', engine_schematics[line_number]):  # fixme
        symbol_pos = match.span()[0]
        symbol_match = match.group()
        current_line['symbol_pos'].append(symbol_pos)
        # print(symbol_pos, symbol_match)
```

```python
def adjacent_symbol_check(integer):
    global gear_sum
    for index, number in enumerate(current_line['number_pos']):
        if any(x + 1 in current_line['symbol_pos'] for x in number) or any(x - 1 in current_line['symbol_pos'] for x in number) is True:
            integer += int(current_line['number_match'][index])
            gear_sum = integer
            print(number, 'true', int(current_line['number_match'][index]))
```

```python
def downwards_symbol_check(integer):
    global gear_sum
    for index, number in enumerate(current_line['number_pos']):
        if any(x in previous_line['symbol_pos'] for x in number) or any(x - 1 in previous_line['symbol_pos'] for x in number) or any(x + 1 in previous_line['symbol_pos'] for x in number) is True:
            matched_number = int(current_line['number_match'][index])
            integer += matched_number
            gear_sum = integer
            print(number, 'true', int(current_line['number_match'][index]))
    print(gear_sum)
```
