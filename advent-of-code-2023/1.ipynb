# Part 1

Opening code list

```python
with open('codes.txt', 'r') as f:
    codes = f.readlines()

codes
```

Removing all non-digits, duplicating codes with only one digit and leaving 2 digits for the rest

```python
two_digit_codes = []

for code in codes:
    only_digits = ''.join(filter(str.isdigit, code)).strip()
    print(only_digits)

    if len(only_digits) < 2:
        only_digits = only_digits + only_digits
    else:
        only_digits = only_digits[0] + only_digits[-1]
    two_digit_codes.append(only_digits)

print('\n')
two_digit_codes = [int(i) for i in two_digit_codes]
print(two_digit_codes)
```

Adding up all numbers

```python
sum(two_digit_codes)
```

# Part 2

mapping out the digits from str to int

```python
word_to_digit = {
    'zero': 0,
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9
}
```

initializing regex and compiling pattern

```python
import re

pattern = re.compile(
    r"(?:one|two|three|four|five|six|seven|eight|nine)", flags=re.M)
```

finding the position of the first and last integer digit

```python
for code in codes:
    code = code.strip()
    print(len(code))
    # This will return a match object for the first digit
    first_match = re.search(r'\d', code)
    # This will return a match object for the last digit in the reversed string
    last_match = re.search(r'\d', code[::-1])
    if first_match and last_match:  # Check if both matches are found
        first_digit = int(first_match.group())  # This will return 7
        last_digit = int(last_match.group())  # This will return 8
        print(
            f'First digit: {first_digit}, position {code.index(str(first_digit))}, last digit: {last_digit} position: {code.index(str(last_digit))}')
    else:
        print('No digits in the string')
```

finding the position of the first and last string digit

```python
for code in codes:
    code = code.strip()
    print(len(code))
    first_match = re.search(pattern, code)
    last_match = re.findall(pattern, code)
    if first_match and last_match:
        first_word = first_match.group()
        last_word = last_match[-1]
        print(f'First word: {first_word}, position {code.index(first_word)}, last word: {last_word} position: {code.index(last_word)}')
    else:
        print('No words in the string')
```

backwards regex matching

```python
code = '459zpnrzbxdsplpspnh9zbqvdnxgdfour2oneightmfd'
code = code[::-1]

backwards_pattern = r'(?:eno|owt|eerht|ruof|evif|xis|neves|thgie|enin)'

match = re.search(backwards_pattern, code) # This will return a match object for 'thgiee'
if match:
    print((match.group())[::-1]) # This will print 'thgiee'
else:
    print('No match')
```

```python
two_digit_codes = []

for code in codes:
    code = code.strip()
    print(code)

    # regexing the integers
    first_match = re.search(r'\d', code)
    last_match = re.search(r'\d', code[::-1])
    first_digit = int(first_match.group())
    last_digit = int(last_match.group())

    # regexing the strings
    first_match = re.search(pattern, code)
    last_match = re.search(backwards_pattern, code[::-1])
    if first_match and last_match:
        first_word = first_match.group()
        last_word = last_match.group()[::-1]
    else:
        first_word = 'None'
        last_word = 'None'

    # determining the orders and the first and last characters
    if first_word and last_word != 'None':
        if code.find(str(first_digit)) > code.find((first_word)) and code.rfind(str(last_digit)) < code.rfind((last_word)):
            print('String comes first, string comes last')
            first_word = word_to_digit.get(first_word, -1)
            last_word = word_to_digit.get(last_word, -1)
            code = str(first_word) + str(last_word)
            code = int(code)
            print(code)
        elif code.find(str(first_digit)) < code.find((first_word)) and code.rfind(str(last_digit)) < code.rfind((last_word)):
            print('Integer comes first, string comes last')
            last_word = word_to_digit.get(last_word, -1)
            code = str(first_digit) + str(last_word)
            code = int(code)
            print(code)
        elif code.find(str(first_digit)) < code.find((first_word)) and code.rfind(str(last_digit)) > code.rfind((last_word)):
            print('Integer comes first, integer comes last')
            code = str(first_digit) + str(last_digit)
            code = int(code)
            print(code)
        elif code.find(str(first_digit)) > code.find((first_word)) and code.rfind(str(last_digit)) > code.rfind((last_word)):
            print('String comes first, integer comes last')
            first_word = word_to_digit.get(first_word, -1)
            code = str(first_word) + str(last_digit)
            code = int(code)
            print(code)
    elif first_word and last_word == 'None':
        print('Integer comes first, integer comes last')
        code = str(first_digit) + str(last_digit)
        code = int(code)
        print(code)

    # appending each code to list of two digit codes
    two_digit_codes.append(code)

    print('\n')
    del first_word, last_word, first_digit, last_digit
```
